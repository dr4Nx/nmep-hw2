```python
import h5py
import matplotlib.pyplot as plt

path = '/honey/nmep/medium-imagenet-96.hdf5'

with h5py.File(path, 'r') as hf:
    images = hf['images-train'][:10] 
    labels = hf['labels-train'][:10] 
    
    images = [image.transpose(1, 2, 0) for image in images]

    fig, ax = plt.subplots(2, 5)
    for i in range(2):
        for j in range(5):
            ax[i, j].imshow(images[i * 5 + j])
            ax[i, j].set_title(f"Label: {labels[i * 5 + j]}")
            ax[i, j].axis('off')

    plt.show()
```
